```python
#LAMBDA -->1. A lambda function is a small ananymous function. Anonymous means the function does not have a name .
          #2. Lambda function are useful when a function is needed only once and does not need reused elsewhere

   # Syntax : lambda arguments : expression --> a = lambda x: x ** 2
   #  a= function name
   #  lambda = function keyword
   #  x= argument
   #  ** = expression

   #Interview Definition

#Lambda function is an anonymous, single-expression function defined using the lambda keyword.

square = lambda x:x*x
print(square(5))

tax = lambda x:x*0.03
print(tax(400000))

sum = lambda a,b : a+b
print(sum(16,11))

max = lambda a,b: a if a>b else b
print(max(16,11))

```

```python
#MAP --> the MAP() function applies the same operation to every item in a collection
# Syntax --> map(function, iterable)

numbers= [1,2,3,4,5]
result= map(lambda x:x *2,numbers)
print(list(result))
```

```python
# FILTER --> The filter() function selects only the items that satisfy a condition
# Syntax --> filter(function, iterable)

numbers = [1,2,3,4,5,6]
even_numbers = filter(lambda x: x%2==0,numbers)
print(list(even_numbers))
```

```python
#REDUCE FUNCTION -->Reduce combines all values into a singel value.
                    #It repeatedly performs an action until only one result remains
#IMPORT --> from functools import reduce
#SYNTAX --> reduce(function , iterable)
from functools import reduce
numbers = [1,2,3,4]
result = reduce(lambda x,y:x+y,numbers)
print(result)
```

```python
#ZIP FUNCTION -->The zip() function combines two or more iterables (lists, tuples, etc.) element by element into pairs (or tuples).
#Syntax --> zip(iterable1,iterable2,...)
names = ["Mukesh", "Rahul", "Kiran"]
marks = [85, 90, 88]

result = list(zip(names, marks))

print(result)
```

```python
#ENUMERATE FUNCTION --> The enumerate() function automatically adds an index number.

fruits = ["apple", "banana", "mango"]
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

```python
#SORTED FUNCTION --> Sort() function sorts the data in ascending or decending order

numbers = [1,3,4,2,7,0]
numbers.sort()
print(numbers)
```

```python
#ANY FUNCTION --> Prints TRUE if any condition is true
a=30
b=45
numbers = [a>b , a<b]
print(any(numbers))
```

```python
#ALL FUNCTION -->The all() function returns True only if all elements in an iterable are True

a=45
b=45
numbers = [a>=b,a<=b,a==b]
print(all(numbers))
```

# SLICING
    Slicing is used to extract a portion of an array.

  Syntax -->
        arr[start : stop : step]

  2D array -->
    Syntax -->
       arr[start_row : end_row : step , start_col : end_col : step ]

```python
import numpy as np

arr = np.arange(10)

print(arr)

print(arr[1:6])

print(arr[::-1])

print(arr[1: :2])

print(arr[1:6:2])

print(arr[:8:3])

a = np.arange(8).reshape(4,2)

print(a)

print(a[0:3,0:1])

print(a[0])

print(a[0,:])
```

```python
import numpy as np
arr= np.arange(10)
print(arr[arr%2==0])

print(arr[(arr>2) & (arr<9)])
```

# BROADCASTING
  --> It allows array to different shapes to perform arithmetic opeations without explicitly reshaping them.
  --> Broadcasting automatically expands the smaller array to match the shape of the larger array.

```python
import numpy as np
a = np.array([1,2,5,8])
print(a+5)

b= np.array([[1,5,10,15],[2,4,6,8]])
print(b+a)

c= np.array([[1],[2]])

print(b+c)
```

# DESCRIPTIVE METHODS
    --> Descriptive methods are used to summerize and analyze data, Providing key statistics like mean, median, standard deviation etc..
    -->They are essential for data analysis, making it easier to spot trends and patterns in a dataset.

  1. sum() --> computes the sum of array elements --> np.sum(arr,axis=0/1/none)
  2. mean() --> Computes the average value of an array.
  3. standard deviation() --> Standard Deviation measures how far the data values are spread from the mean.
      --> Low standard deviation → Data points are close to the mean.
      --> High standard deviation → Data points are spread far from the mean.
  4. Variance() --> Variance measures how far data values are spread from the mean (average).
      --> It tells us the average of the squared differences from the mean.

          --> Small variance → Data is close to the mean.
          --> Large variance → Data is widely spread.
  5. Minimum() --> Returns minimum value of the array
  6. Maximum() --> Returns maximum value of the array
  7. argmin() --> Return the index of the minimum value of the array
  8. argmax() --> Returns teh index of the maximun value of the array
  


```python
import numpy as np
#SUM
a=np.arange(1,5)
# a=np.array([1,2,3,4,5])
print(np.sum(a))

b=np.arange(1,9).reshape(2,4)
print(b)
print(np.sum(b))
print(np.sum(b,axis=0))
print(np.sum(b,axis=1))

#MEAN
print(np.mean(a))
print(np.mean(b))
print(np.mean(b,axis=0))
print(np.mean(b,axis=1))

#STANDARD DEVIATION
print(np.std(a))

#VARIANCE
print(np.var(a))
print(np.var(b))
print(np.var(b,axis=0))
print(np.var(b,axis=1))

#MINIMUM
print(np.min(a))
print(np.min(b))
print(np.min(b,axis=0))
print(np.min(b,axis=1))

#MAXIMUM
print(np.max(a))
print(np.max(b))
print(np.max(b,axis=0))
print(np.max(b,axis=1))

#ARGMIN
c=np.array([8,4,52,1,0])
print(np.argmin(c))
print(np.argmin(b))
print(np.argmin(b,axis=0))
print(np.argmin(b,axis=1))

#ARGMAX
print(np.argmax(c))
print(np.argmax(b))
print(np.argmax(b,axis=0))
print(np.argmax(b,axis=1))
```
