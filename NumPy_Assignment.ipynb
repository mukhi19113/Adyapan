```python
import numpy as np
```

1. The Clean Slate: Create a 1D array of numbers from 0 to 9, and then reverse it.

```python
np.arange(10)[::-1]
```

2. The Grid: Create a 3x3 matrix with values ranging from 0 to 8

```python
np.arange(0,9).reshape(3,3)
```

3. The Replacer: Create a 1D array of numbers from 1 to 10. Replace all odd numbers
with -1 without changing the original array.

```python
a= np.arange(1,11)
a[a%2==1]= -1
print(a)
```

4. The Boundary: Create a 5x5 matrix with all 1s on the border and 0s on the inside

```python
arr = np.ones((5,5)).astype(int)
arr[1:-1,1:-1]=0
print(arr)
```

5. The Common Ground: Find the common elements between two NumPy arrays: a =
np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6]) and b = np.array([7, 2,
10, 2, 7, 4, 9, 4, 9, 8]).

```python
a= np.array([1,2,3,2,3,4,3,4,5,6])
b= np.array([7,2,10,2,7,4,9,4,9,8])
common_elements = np.intersect1d(a,b)
common_elements
```

6. The Filter: Extract all numbers between 5 and 10 (inclusive) from this array: a =
np.array([2, 6, 1, 9, 10, 3, 27, 8, 4]).

```python
a= np.array([2,6,1,9,10,3,27,8,4])
print(a[(a>5) & (a<=10)])
```

7. The Matchmaker: Given two arrays a and b of the same shape, get the positions
(indices) where elements of a and b match.

```python
a= np.array([1,2,3,4,3,4])
b= np.array([4,5,2,6,7,4])
np.where(a==b)[0]
```

8. The Normalizer: Create a random 5x5 matrix and normalize it (so that all values scale
between 0 and 1 using min-max normalization).

```python
a= np.random.random((5,5))
normalized_array = (a-a.min())/(a.max()-a.min())
normalized_array
```

9. The Stack Overflow: Stack two arrays a = np.arange(10).reshape(2,5) and b
= np.repeat(1, 10).reshape(2,5) vertically.

```python
a=np.arange(10).reshape(2,5)
b= np.repeat(1,10).reshape(2,5)
np.vstack((a,b))
```

10. The Peak Finder: Find the maximum value in each row of a 2D random 4x4 array, and
return both the values and their column indices.

```python
a= np.random.random((4,4))
print(np.max(a,axis=1))
print(np.argmax(a,axis=1))

```
