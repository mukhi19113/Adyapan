10. Join the Arrays:
Learn how to concatenate two 1D arrays both vertically and horizontally. This operation is critical in combining data from different sources.

```python
a=np.arange(10)
b=np.arange(10,20)
concat_array = np.concatenate((a,b))
concat_array
```

```python
a=np.arange(6).reshape(2,3)
b=np.arange(10,16).reshape(2,3)
concat_array1 = np.concatenate((a,b),axis= 1) # horizontal
concat_array2 = np.concatenate((a,b),axis = 0) # Vertical
print(concat_array1)
print(concat_array2)
```

11. Common Ground:
Find the intersection of two one-dimensional arrays. This exercise is useful in data comparison tasks, such as finding common elements in data sets.

```python
arr1 = np.arange(1,7)
arr2 = np.array([1,8,3,9,2])
common_elements = np.intersect1d(arr1,arr2)
common_elements
```

```python
arr = np.array([[1,2,5],[3,4,2]])
arr2 = np.array([[2,5,3],[7,9,3]])
common_elements = np.intersect1d(arr,arr2)
common_elements
```
